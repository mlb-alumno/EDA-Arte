# Avanzando y conluyendo la hipótesis: palabras clave en las descripciones de los eventos

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import re
import nltk
from nltk.tokenize import RegexpTokenizer
import plotly.express as px

df_casa = pd.read_csv("../data/casa-encendida/casa_enc_sep.csv",index_col=0).drop("Unnamed: 0",axis=1)
```

```python
# Definimos las funciones necesarias para procesar los textos
def conteo_palabras(text,lista_stop_words):
    palabras = RegexpTokenizer(r'\w+').tokenize(text)
    stopwords = nltk.corpus.stopwords.words('spanish')
    stopwords.extend(('la','el','en','años','the','su','ser','parte','través','tal','cual')) 
    stopwords.extend(lista_stop_words)
    conteo = nltk.FreqDist(w.lower() for w in palabras if w.lower() not in stopwords)
    return pd.DataFrame(conteo.items(), columns=['palabra', 'frecuencia']).sort_values("frecuencia",ascending=False).reset_index(drop=True)
def texts_per_year(num_year,filter="description"):
    df_casa_sinnan = df_casa[df_casa["year"] == num_year].copy()
    df_casa_sinnan = df_casa_sinnan[df_casa_sinnan[filter].notna()]
    return ' '.join(df_casa_sinnan[filter].values.tolist())
```

## Bigramas y trigramas para mayor contexto de las palabras

```python
from nltk.collocations import *
# Buscamos los Bigramas que más aparecen
def bigrams(text,searched_word):
    tokens = RegexpTokenizer(r'\w+').tokenize(text)
    mis_stop_words = ('.',',')
    finder = BigramCollocationFinder.from_words(tokens)
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    finder.apply_ngram_filter(lambda *w: searched_word not in w)
    finder.apply_ngram_filter(lambda *w: w in mis_stop_words)
    list = sorted(finder.nbest(bigram_measures.likelihood_ratio, 20))
    result_list = []
    for i in range(len(list)):
        result_list.append(' '.join(list[i]))
    return result_list
# Buscamos los Trigramas que más aparecen
def trigrams(text,searched_word,nwords=20):
    tokens = RegexpTokenizer(r'\w+').tokenize(text)
    mis_stop_words = ('.',',')
    finder = TrigramCollocationFinder.from_words(tokens)
    trigram_measures = nltk.collocations.TrigramAssocMeasures()
    finder.apply_ngram_filter(lambda *w: searched_word not in w)
    finder.apply_ngram_filter(lambda *w: w in mis_stop_words)
    list = sorted(finder.nbest(trigram_measures.likelihood_ratio, nwords))
    result_list = []
    for i in range(len(list)):
        result_list.append(' '.join(list[i]))
    return result_list
```

```python
import plotly.express as px
def trigram_per_year(year,n_words=10,n_trigrams=10):
    filtro = ['cómo','cada','puede','mejor','si','así','casa','encendida','five','give','dos','1','30','horas','2','3','hora','minutos','ciclo','sesión','duración','45','sesion']
    tree_conteo_df = conteo_palabras(texts_per_year(year),filtro)[0:n_words].copy()
    df_trigram = pd.DataFrame()
    for i in tree_conteo_df["palabra"].values:
        df_pr = pd.DataFrame({'palabra':i,'frecuencia':tree_conteo_df[tree_conteo_df["palabra"]==i]["frecuencia"].values[0],'trigram':trigrams(texts_per_year(year),i,n_trigrams),'year':year})
        df_trigram = pd.concat([df_trigram,df_pr]).reset_index(drop=True)
    return df_trigram
def all_trigrams(n_word=10,n_trigram=10):
    all_trigrams_df = pd.DataFrame()
    for i in range(2012,2024):
        df_pr = trigram_per_year(i,n_words=n_word,n_trigrams=n_trigram)
        all_trigrams_df = pd.concat([all_trigrams_df,df_pr]).reset_index(drop=True)
    return all_trigrams_df


```

## Visualizamos las palabras con más apariciones por año y sus trigramas asociados

```python
fig = px.treemap(all_trigrams(), path=[px.Constant("Palabras con más apariciones"),'year','palabra','trigram'],values='frecuencia',labels=conteo_2020.columns)
fig.update_traces(root_color="lightgrey")
fig.update_traces(
    hovertemplate='<b></b>' 
)
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
fig.write_html("../img/12-palabras.html")
```
